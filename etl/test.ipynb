```python
import mysql.connector
import pandas as pd
import glob
import numpy as np
import os
import csv
import yaml
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse


df2 = pd.read_csv('trend1.csv', parse_dates=['date'], index_col='date')

df2.drop(columns = ['business','Unnamed: 0'], inplace = True)

```

```python
df2
```

```python
result_mul = seasonal_decompose(df2['sales'], model='multiplicative', extrapolate_trend='freq')
plt.rcParams.update({'figure.figsize': (10,10)})
result_mul.plot().suptitle('Multiplicative Decompose', fontsize=22)
```
